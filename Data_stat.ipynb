```python
import pandas as pd
import os
from IPython.display import display
```

```python
downloads_path = os.path.expanduser('~/Downloads')
csv_files = ['events_header.csv','users.csv','campaigns.csv']
```

```python
all_dataframes = {}
for file in csv_files:
    file_path = os.path.join(downloads_path, file)
    df = pd.read_csv(file_path)
    display(df.head(3))
    print("\n")
```

```python
dataframes = {}

for file in csv_files:
    file_path = os.path.join(downloads_path, file)

    print(f"\n{'='*50}")
    print(f"АНАЛІЗ ФАЙЛУ: {file}")
    print(f"{'='*50}\n")
    df = pd.read_csv(file_path)
    dataframes[file] = df
    # Збираємо статистику по кожній колонці
    stats = []
    for col in df.columns:
        stats.append({
            'Назва колонки': col,
            'Формат даних': df[col].dtype,
            'К-сть значень (не порожніх)': df[col].count(),
            'К-сть порожніх (NaN)': df[col].isnull().sum(),
            'К-сть унікальних': df[col].nunique()
        })

    # Формуємо та виводимо таблицю зі статистикою
    stats_df = pd.DataFrame(stats)
    print(f"Детальна статистика по колонках для {file}:")
    display(stats_df)
```
